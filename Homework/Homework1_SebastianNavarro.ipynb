### 1. Assume we implemented an algorithm that has complexity approximately $nlog(n)$ . How much faster is this implementation compared to one of complexity $n^2$ 


```python
from time import time
import numpy as np
import pandas as pd
from time import sleep
import matplotlib.pyplot as plt

x_numbers = range(100,131)
nlogn_values = []
nlogn_time = []
n2_values=[]
n2_time = []

def timer_func(func):
    # This function shows the execution time of the function object passed
    def wrap_func(*args, **kwargs):
        t1 = time()*1000
        result = func(*args, **kwargs)
        t2 = (time()*1000)-1000
        return round(result,2),round(t2-t1,2)
    return wrap_func

@timer_func
def fnlogn(n):
    sleep(1)
    return n*np.log(n)

@timer_func
def fn2(n):
    sleep(1)
    return pow(n,2)

for x in x_numbers:
    nlogn_values.append(fnlogn(x)[0])
    nlogn_time.append(fnlogn(x)[1])
    n2_values.append(fn2(x)[0])
    n2_time.append(fn2(x)[1]) 

df = pd.DataFrame(zip(x_numbers,nlogn_values,nlogn_time,n2_values,n2_time))
df.columns = ['n','n log (n)','Time','n^2','Time']
df
```

```python
plt.plot(x_numbers, nlogn_time, label='nlog(n)')
plt.plot(x_numbers, n2_time, label='n^2')
plt.legend()
plt.title("Big O-Complexity")
plt.show()
```

Por tanto, se observa que $n^2$ crece mucho más rápido que $nlog(n)$. Entonces un algoritmo implementado con $nlog(n)$ es mucho más rápido.

### 2.The runtime of an algorithm is captured in the following table:

```python
import pandas as pd
import matplotlib.pyplot as plot
N = [1000, 2000, 4000, 8000, 16000, 32000, 64000]
n = [0.0, 0.02, 0.20, 0.60, 2.60, 10.4, 41.6]

df = pd.DataFrame(zip(N, n))
df.columns = ['n', 'T(n)']
df
```

1. As precissely as possible, estimate the function that describes the growth of the runtime

Dado que se asemeja a una ecuación cuadrática, podemos usar la siguiente fórmula:

$${y = a(x-h)^2 + k}$$ 

La función comienza en cero, que corresponde al vértice, entonces:
$${v=(h=0 , k=0)=(0,0)}$$

Tomando valores de la tabla:
$${n=x=16000}$$
$${T(n)=y=2.60}$$

Reemplazando:
$${2.60 = a(16000-0)^2 + 0}$$ 
$${a = 1.0156 \times 10^{-8}}$$

Finalmente, reemplazando a en la fórmula original:
$$T(n)=1.0156 \times 10^{-8}n^2$$

Con esta ecuación estimada, calculemos los valores que toma la función para $n$:

```python
import pandas as pd
import matplotlib.pyplot as plot

N=[1000, 2000, 4000, 8000, 16000, 32000, 64000]
n = []

def f(n):
    return 1.0156E-08*pow(n,2)

for x in N:
    n.append(round(f(x),2))


df = pd.DataFrame(zip(N, n))
df.columns = ['n', 'T(n)']
df
```

```python
df.plot(x='n', y='T(n)')
```

2. Use Python time library to estimate the runtime for inputs of $n=[1,10,10^2,10^3,10^4,10^5,10^6,10^7,10^8,10^9]$

```python
from time import time
import numpy as np
import pandas as pd
from time import sleep
import matplotlib.pyplot as plt

x_numbers = [1,10,pow(10,2),pow(10,3),pow(10,4),pow(10,5),pow(10,6),pow(10,7),pow(10,8),pow(10,9)]
sum1_values = []
sum1_time = []
sum2_values=[]
sum2_time = []

def timer_func(func):
    # This function shows the execution time of the function object passed
    def wrap_func(*args, **kwargs):
        t1 = time()*1000
        result = func(*args, **kwargs)
        t2 = (time()*1000)-1000
        return round(result,2),round(t2-t1,2)
    return wrap_func


@timer_func
def sum1(n):
    sleep(1)
    total = 0
    for i in range(1, n+1):
        total += i
    return total

@timer_func
def sum2(n):
    sleep(1)
    total = n*(n+1)//2
    return total

print(sum1(5)[0],' = ', sum2(5)[0])
print(sum1(8)[0],' = ', sum2(8)[0])
print(sum1(103)[0],' = ', sum2(103)[0])
print(sum1(527)[0],' = ', sum2(527)[0])

for x in x_numbers:
    sum1_values.append(sum1(x)[0])
    sum1_time.append(sum1(x)[1])
    sum2_values.append(sum2(x)[0])
    sum2_time.append(sum2(x)[1])
    
df = pd.DataFrame(zip(x_numbers,sum1_values,sum1_time,sum2_values,sum2_time))
df.columns = ['n','T1(n)','Time','T2(n)','Time']
df
```

```python
plt.plot(x_numbers,sum1_time,label='T1(n)')
plt.plot(x_numbers,sum2_time,label='T2(n)')
plt.legend()
plt.show()
```
