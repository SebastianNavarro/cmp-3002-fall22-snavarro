### 1. Assume we implemented an algorithm that has complexity approximately $nlog(n)$ . How much faster is this implementation compared to one of complexity $n^2$ 


```python
from time import time
import numpy as np
import pandas as pd
from time import sleep
import matplotlib.pyplot as plt

x_numbers = range(100,131)
nlogn_values = []
nlogn_time = []
n2_values=[]
n2_time = []

def timer_func(func):
    # This function shows the execution time of the function object passed
    def wrap_func(*args, **kwargs):
        t1 = time()*1000
        result = func(*args, **kwargs)
        t2 = (time()*1000)-1000
        return round(result,2),round(t2-t1,2)
    return wrap_func

@timer_func
def fnlogn(n):
    sleep(1)
    return n*np.log(n)

@timer_func
def fn2(n):
    sleep(1)
    return pow(n,2)

for x in x_numbers:
    nlogn_values.append(fnlogn(x)[0])
    nlogn_time.append(fnlogn(x)[1])
    n2_values.append(fn2(x)[0])
    n2_time.append(fn2(x)[1]) 

df = pd.DataFrame(zip(x_numbers,nlogn_values,nlogn_time,n2_values,n2_time))
df.columns = ['n','n log (n)','Time','n^2','Time']
df
```

```python
plt.plot(x_numbers, nlogn_time, label='nlog(n)')
plt.plot(x_numbers, n2_time, label='n^2')
plt.legend()
plt.title("Big O-Complexity")
plt.show()
```

Por tanto, se observa que $n^2$ crece mucho ms rpido que $nlog(n)$. Entonces un algoritmo implementado con $nlog(n)$ es mucho ms rpido.

```python
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
class Singly_linked_list:
    """
    Implementation of a singly linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
        
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node

class Singly_linked_list(Singly_linked_list):        
    def delete_head(self):
        # delete to the head
        # A -> B -> null
        # R -> B -> null
        self.head_node = self.head_node.next_node
            
    def delete_middle(self,value):
        # A -> B -> C -> null
        # A -> C -> null
        prev = self.head_node
        node = prev.next_node
        
        if prev.val == value:
            self.head_node = self.head_node.next_node
            prev.next_node = None
        
        else:
            while node != value:
                prev = node
                node = node.next_node
                
                if node==None:
                    break
                
            if node:
                prev.set_next_node(node.next_node)
                node.next_node=None
            
            else:
                print("No se encuentra")
                
    def delete_tail(self):
        # delete to the tail
        # A -> B -> null
        # A -> R -> null 
        prev=self.head_node
        node = prev.next_node
        while node.next_node!=None:
            prev=node
            node = node.next_node   
        prev.set_next_node(node.next_node)
```

```python
import ctypes

class Array(object):
    """
    Implementation of the array data structure
    """

    def __init__(self, n):
        """
        Initialize the class
        """
        self.l = 0
        self.n = n
        self.array = self._create_array(self.n)        
    
    def _create_array(self, n):
        """
        Creates a new array of capacity n
        """
        return (n * ctypes.py_object)()

class Array(Array):
    def __getitem__(self, item_index):
        """
        Return element at item_index
        """
        if (item_index < 0) or (item_index >= self.n):
            raise IndexError('index out of range!')
        try:
            x = self.array[item_index]
        except ValueError:
            x = None
        return x
    
    def __setitem__(self, item_index, item):
        """
        Set element at item_index
        """
        if (item_index < 0) or (item_index >= self.n):
            raise IndexError('index out of range!')
        self.array[item_index] = item

    def __init__(self, n, values=None):
        self.l = 0
        self.n = n
        self.array = self._create_array(self.n)
        if values:
            self.initialize_array(values)
            
    def initialize_array(self, values):
        """
        Initialize array
        """
        if self.n != len(values):
            raise ValueError("element count different than capacity")
        for item in values:
            self.array[self.l] = item
            self.l += 1
            
    def list_array(self):
        """
        List elements of the array
        using list comprehension
        """
        return ", ".join(str(x) if x is not None else '_' for x in self)

    def list_array2(self):    
        y = []    
        for x in self:
            if x is not None:
                y.append(str(x))
            else:
                y.append('_')
        return ", ".join(y)

    def insert(self, index, element):
        """
        implementation of insert
        """
        if self.l == self.n:
            raise ValueError("no more capacity")
        if (index < 0) or (index > self.l):
            raise IndexError('index out of range!')
        x = self.l
        while x > index:
            self.array[x] = self.array[x-1]
            x -= 1
        self.array[index] = element
        self.l += 1

    def insert_to_head(self, item):
        """
        Add new item to the beginning of the array
        """
        i = self.l-1
        while (i > 0):
            self.array[i] = self.array[i-1]
            i -= 1
        self.array[0] = item
        self.l += 1
        
    def delete_to_tail(self):
        self.array[self.l-1]='_'
        self.l -= 1
    
    def delete_to_head(self):
        i=1
        while (i > 0 and i<self.l):
            self.array[i-1] = self.array[i]
            i += 1
        self.l -= 1
        self.array[self.l] = '_'     
    
    def delete(self, index):
        i = self.l-1
        j=index
        while i > index:
            self.array[j] = self.array[j+1]
            i -= 1
            j += 1
        self.array[self.l-1] = '_'
        self.l -= 1   
        
A=Array(5)
A.insert(0, 0)
A.insert(1, -2)
A.insert(2, 4)
A.insert(3, 5)
A.insert(4, 3)
print(A.list_array())
A.delete(2)
print(A.list_array())
A.delete_to_head()
print(A.list_array())
A.delete_to_tail()
print(A.list_array())
```
