## 1. Implement the class Stacks and all its methods using singly linked lists. Analyze the runtime and memory complexity, and compare with what we studied in class

```python
import ctypes
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
        
class Stack(object):
    """
    Implementation of the stack data structure
    """

    def __init__(self, n):
        self.head_node = None
        self.l = 0
        self.n = n
        self.stack = self._create_stack(self.n)        
    
    def _create_stack(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()

class Stack(Stack):
    def push(self,stack):
        node = Node (stack)
        if self.l == self.n:
            raise ValueError("no more capacity")
            
        if(self.head_node==None):
            self.head_node = node
        else:
            node.next_node=self.head_node
            self.head_node=node
        self.l += 1
            
    def pop(self):
        if(self.head_node==None):
            raise ValueError('stack is empty')
        else:
            self.head_node=self.head_node.next_node
        self.l -= 1
        return self.head_node
 
    def top(self):
        return self.head_node.val

    def full(self):
        """
        Is the stack full?
        """      
        if self.l == self.n:
            return True
        return False
    
    def empty(self):
        """
        Is the stack empty?
        """
        if self.l == 0:
            return True
        return False
    
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node

            
    def size(self):
        """
        Return size of the stack
        """
        return self.l
        
            
S=Stack(10)
S.push(1)
S.push(2)
S.push(4)
S.push(-1)
print(S.list_traversed())
S.pop()
print(S.size())
print(S.list_traversed())
print(S.full())
print(S.empty())
```

## 4. Complete the PriorityQueue class, so that when we call dequeue, the element with the highest priority will be returned. Analyze the complexity of runtime and memory of the enqueue and dequeue methods.

```python
import ctypes
class PriorityQueue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n)        
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
    def enqueue(self, item):
        """
        Add new item to the queue
        """
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.queue[self.l] = item
        self.l += 1
    
    #primer elemento: Prioridad (menor valor, mayor prioridad)
    #Segundo elemento: Valor
    
    def insert(self, item):
        self.enqueue(item)
        l=self.l
        
        for i in range(1,l-1):
            if(item[0]>=self.queue[i-1][0] and item[0]<=self.queue[i][0]): 
                #print(i)
                for j in range(l-1,i,-1):
                    #print(j)
                    self.queue[j]=self.queue[j-1]   

                self.queue[i]=item
                break
    
    def sort(self):
        for j in range (2, self.l):
            key = self.queue [j]
            i = j - 1
            while i > 0 and self.queue [i] > key:
                self.queue [i + 1] = self.queue [i]
                i = i - 1
            self.queue [i + 1] = key  
    
q = PriorityQueue(10)
q.enqueue((1,5))
q.enqueue((9,2))
q.enqueue((8,4))

q.sort()
print(q.queue[0:q.l])
q.insert((2,3))
print(q.queue[0:q.l])
```
