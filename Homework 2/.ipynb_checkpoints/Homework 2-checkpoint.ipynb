## 1. Implement the class Stacks and all its methods using singly linked lists. Analyze the runtime and memory complexity, and compare with what we studied in class

```python
import ctypes
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val):
        self.val = val
        self.next_node = None
        
class Stack(object):
    """
    Implementation of the stack data structure
    """

    def __init__(self, n):
        self.head_node = None
        self.l = 0
        self.n = n
        self.stack = self._create_stack(self.n)        
    
    def _create_stack(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()

class Stack(Stack):
    def push(self,stack): #Complexity O(1)
        node = Node (stack)
        if self.l == self.n:
            raise ValueError("no more capacity")
            
        if(self.head_node == None):
            self.head_node = node
        else:
            node.next_node = self.head_node
            self.head_node = node
        self.l += 1
            
    def pop(self): #Complexity O(1)
        if(self.head_node == None):
            raise ValueError('stack is empty')
        else:
            node = self.head_node
            self.head_node = self.head_node.next_node        
        self.l -= 1
        return node
 
    def top(self): #Complexity O(1)
        return self.head_node.val

    def full(self): #Complexity O(1)
        """
        Is the stack full?
        """      
        if self.l == self.n:
            return True
        return False
    
    def empty(self): #Complexity O(1)
        """
        Is the stack empty?
        """
        if self.l == 0:
            return True
        return False
    
    def size(self): #Complexity O(1)
        """
        Return size of the stack
        """
        return self.l
    
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node
   
S=Stack(10)
S.push(1)
S.push(2)
S.push(4)
S.push(-1)
print("Stack printed:")
S.list_traversed()
print("Stack has a size of: " + str(S.size()))
print("Element removed from the Stack: " + str(S.pop().val))
print("Stack has a size of: " + str(S.size()))
print("\nNew Stack printed:")
S.list_traversed()
print("Is Stack full? " + str(S.full()))
print("Is Stack empty? " + str(S.empty()))
print("Element removed from the Stack: " + str(S.pop().val))
print("Element removed from the Stack: " + str(S.pop().val))
print("Element removed from the Stack: " + str(S.pop().val))
print("Is Stack empty? " + str(S.empty()))
```

## 2. Write a method part of the linked list class that will reverse the linked list. Your implementation should visit each node in the list only one time, and should use  $O(1)$ of extra memory. 

```python
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
class Singly_linked_list:
    """
    Implementation of a singly linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
        
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node

class Singly_linked_list(Singly_linked_list):        
    def delete_head(self):
        # delete to the head
        # A -> B -> null
        # R -> B -> null
        self.head_node = self.head_node.next_node
        
    def delete_tail(self):
        # delete to the tail
        # A -> B -> null
        # A -> R -> null 
       prev=self.head_node
       node = prev.next_node
       while node.next_node!=None:
             prev=node
             node = node.next_node   
       prev.set_next_node(node.next_node)
            
    def delete_middle(self,value):
        # A -> B -> C -> null
        # A -> C -> null
        prev=self.head_node
        node = prev.next_node
        
        if prev.val == value:
            self.head_node=self.head_node.next_node
            prev.next_node=None
        
        else:
            while node != value:
                prev = node
                node = node.next_node	
                
                if node==None:
                    break
                
            if node:
                prev.set_next_node(node.next_node)
                node.next_node=None
            
            else:
                print("No se encuentra")

    def reverse(self): #Complexity O(1)
        new_node = self.head_node
        prev = None
        
        while new_node:         
            next_node = new_node.next_node
            new_node.next_node =prev
            prev=new_node
            new_node=next_node
        self.head_node=prev
        
m1 = Node("A")
m2 = Node("B")
m3 = Node("C")
m4 = Node("D")

m1.set_next_node(m2)
m2.set_next_node(m3)
m3.set_next_node(m4)

list1 = Singly_linked_list(m1)
print("List 1")
list1.list_traversed()
list1.reverse()
print("List 1 reversed")
list1.list_traversed()
```

## 3. Implement the class Queue using stacks.


```python
import ctypes
class Queue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n) 
        self.stack_1=[]
        self.stack_2=[]
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
class Queue(Queue):
    def enqueue(self, item):
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.stack_1.append(item)
        self.l += 1
        print("Element added to the Queue: " + str(item))
        
    def dequeue(self):
 
        # if both the stacks are empty
        if len(self.stack_1) == 0 and len(self.stack_2) == 0:
            raise ValueError('stack is empty')
 
        # if s2 is empty and s1 has elements
        if len(self.stack_2)==0:
            while self.stack_1:
                self.stack_2.append(self.stack_1.pop())
        self.l -= 1
        return self.stack_2.pop()

    def first(self):
        """
        Show the first element of the queue
        """
        if(len(self.stack_1)==0):
            return self.stack_2[len(self.stack_2)-1]
        else:
            return self.stack_1[0]
    
    def full(self):
        """
        Is the queue full?
        """
        if self.l == self.n:
            return True
        return False     

    def empty(self):
        """
        Is the queue empty?
        """
        if self.l == 0:
            return True
        return False

    def size(self):
        """
        Return size of the queue
        """
        return self.l
    
q = Queue(3)
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

print('Is Queue empty? ' + str(q.empty()))
print('Queue has a size of: ' + str(q.size()))
print('Is Queue full? ' + str(q.full()))
print('Queue s first item is: ' + str(q.first()))
print("Element removed from the Queue: " + str(q.dequeue()))
print('Queue s first item is ' + str(q.first()))
print("Element removed from the Queue: " + str(q.dequeue()))
print("Element removed from the Queue: " + str(q.dequeue()))
print('Is Queue empty? ' + str(q.empty()))
print('Queue has a size of: ' + str(q.size()))
print('Is Queue full? ' + str(q.full()))          
```

## 4. Complete the PriorityQueue class, so that when we call dequeue, the element with the highest priority will be returned. Analyze the complexity of runtime and memory of the enqueue and dequeue methods.

```python
import ctypes
class PriorityQueue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n)        
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
    def enqueue(self, item):
        """
        Add new item to the queue
        """
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.queue [self.l] = item
        self.l += 1
        
    def dequeue(self):
        self.sort()
        return self.queue [0]
    
    #primer elemento: Prioridad (menor valor, mayor prioridad)
    #Segundo elemento: Valor
    
    def insert(self, item):
        self.enqueue(item)
        l = self.l
        for i in range(1, l-1):
            if(item [0] >= self.queue [i-1] [0] and item [0] <= self.queue [i] [0]): 
                for j in range (l-1, i,-1):
                    self.queue [j] = self.queue [j-1]   
                self.queue [i] = item
                break
    
    def sort(self):
        for j in range (2, self.l):
            key = self.queue [j]
            i = j - 1
            while i > 0 and self.queue [i] > key:
                self.queue [i + 1] = self.queue [i]
                i = i - 1
            self.queue [i + 1] = key  
    
q = PriorityQueue(10)
q.enqueue((2,5))
q.enqueue((9,2))
q.enqueue((8,4))

print("Queue sorted:")
q.sort()
print(q.queue[0:q.l])
print("\nQueue inserted (2,3):")
q.insert((2,3))
print(q.queue[0:q.l])
print("\nQueue s element with highest priority:")
print(q.dequeue())
```

### 5. A given linked-list (singly or doubly) represents an integer number. For example, 345 is represented by the singly-linked list 3 -> 4 ->5. 

```python
import random
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None,next_node=None):
        self.val = val
        self.next_node=next_node
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
class Singly_linked_list:
    """
    Implementation of a singly linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
            
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val, end =' ')
            node = node.next_node
            
    def insert(self, new_node):
        current = self.head_node
        if current:
            while current.next_node:
                current = current.next_node
            current.next_node = new_node
        else:
            self.head_node = new_node
            
    def reverse(self):
        new_node = self.head_node
        prev = None
        
        while new_node:         
            next_node = new_node.next_node
            new_node.next_node =prev
            prev=new_node
            new_node=next_node
        self.head_node=prev
        
        
def transform(num):
    list0=Singly_linked_list()
    for x in str(num):
        list0.insert(Node(int(x)))
    return list0

def sum_list(list1,list2,list3):
    list1.reverse()
    list2.reverse()
    list3.reverse()

    node_1=list1.head_node
    node_2=list2.head_node
    node_3=list3.head_node
    st=''
    res=0
    
    while node_1:
       
        sum_1=sum([node_1.val,node_2.val,node_3.val])
        
        if(sum_1+res >= 10) and not (node_1.next_node==None):
            st=st+str(sum_1+res)[1]
            res=int(str(sum_1+res)[0])
            
        elif(sum_1+res < 10) and not (node_1.next_node==None):
            st=st+str(str(sum_1+res))
            res=0
            
        elif(node_1.next_node==None):
            st=st+str(str(sum_1+res))[::-1]
            
        node_1=node_1.next_node
        node_2=node_2.next_node
        node_3=node_3.next_node
        
    ans=transform(st)
    ans.reverse()
    return ans

def generate_number(num):
    return random.randint(10**(num-1), 10**(num))

print("List 1:")
list1=transform(345)
list1.list_traversed()
print("\nList 2:")
list2=transform(123)
list2.list_traversed()
print("\nList 3:")
list3=transform(567)
list3.list_traversed()
print("\nList 1 + List 2 + List 3:")
list4=sum_list(list1,list2,list3)
list4.list_traversed()

print("\n\nFor random number of digits:")
print("List 1:")
list5=transform(generate_number(70))
list5.list_traversed()
print("\nList 2:")
list6=transform(generate_number(70))
list6.list_traversed()
print("\nList 3:")
list7=transform(generate_number(70))
list7.list_traversed()
print("\nList 1 + List 2 + List 3:")
list8=sum_list(list5,list6,list7)
list8.list_traversed()
```
